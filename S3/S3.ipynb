### 1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU 

```python
def CpuComp(x):
    return x.split()[0]

df['Cpu_Company'] = df['Cpu'].apply(CpuComp)
df.head()
```

### 1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
def mem_num (s):
    if 'TB' in s.split(' ')[0]:
        return int(float(s.split(' ')[0].strip().replace('TB', ''))*1024)
    else: return int(float(s.split(' ')[0].strip().replace('GB', '')))

df['Memory_Amount'] = df['Memory'].apply(mem_num)
```

```python
df[df['Ram'] == '12GB' ]
```

```python
df['Memory_Amount'].value_counts()
```

### 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD, SDD и т.д.)

```python
def mem_type (s):
    s1 = s.split('+')[0]
    return s1.split(' ',1)[1].strip()

df['Memory_Type'] = df['Memory'].apply(mem_type)
```

```python
df.head()
```

```python
df['Memory_Type'].value_counts()
```

### 1.4 Удалите признаки Memory и ScreenResolution 

```python
df.drop(columns=['Memory', 'ScreenResolution', 'Unnamed: 0'], inplace=True)
```

### 5.1 Ноутбуков каких компаний и с каким процессором больше?	

```python
df.pivot_table(index='Company',
                  columns='Cpu_Company', 
                  values='Price_euros',
                  aggfunc='count',
                  margins=True,
                  margins_name='Total').sort_values('Total', ascending=False)
```

### 5.2 С каким типом памяти и с каким объемом памяти больше ноутбуков? 

```python
df.pivot_table(index='Memory_Type',
                  columns='Memory_Amount', 
                  values='Price_euros',
                  aggfunc='count',
                  margins=True,
                  margins_name='Total').sort_values('Total', ascending=False)
```
