### 5.1 Выберите ноутбук клиенту				
Клиент хочет подобрать ноутбук с 8GB или 16GB ОЗУ на Windows 10 в стоимости до 500 евро, сколько у него вариантов?

```python
ds[(ds['Gpu'] == 'Nvidia GeForce GTX 1050 Ti') & (ds['Company'] == 'MSI') & (~(ds['Inches'] == 15.6))]
```

### 5.3 Что дешевле?				
В среднем дешевле ноутбуки с CPU Intel Core i7 7700HQ 2.8GHz или с Intel Core i7 7600U 2.8GHz? 

```python
print('Intel Core i7 7700HQ 2.8GHz = ',ds[(ds['Cpu'] == 'Intel Core i7 7700HQ 2.8GHz')]['Price_euros'].mean())
print('Intel Core i7 7600U 2.8GHz = ', ds[(ds['Cpu'] == 'Intel Core i7 7600U 2.8GHz')]['Price_euros'].mean())
```
